# LazyQML Intro

This notebook shows the current public API with `Embedding.ZZ` and `Embedding.ZZ_LOCAL`.
`ZZ` is all-to-all; `ZZ_LOCAL` is nearest-neighbour.


```python
from sklearn.datasets import load_iris
from lazyqml import QuantumClassifier
from lazyqml.Global import Embedding, Model
from lazyqml.Utils import set_simulation_type

X, y = load_iris(return_X_y=True)
set_simulation_type("statevector")
```

```python
classifier = QuantumClassifier(
    nqubits={4},
    classifiers={Model.QSVM},
    embeddings={Embedding.ZZ, Embedding.ZZ_LOCAL},
)

scores = classifier.fit(X, y, test_size=0.3, showTable=True)
scores["summary"]
```

```python
set_simulation_type("tensor")

tensor_classifier = QuantumClassifier(
    nqubits={4},
    classifiers={Model.MPSQSVM},
    embeddings={Embedding.ZZ, Embedding.ZZ_LOCAL},
)

tensor_scores = tensor_classifier.repeated_cross_validation(
    X,
    y,
    n_splits=3,
    n_repeats=1,
    showTable=True,
)
tensor_scores["summary"]
```
